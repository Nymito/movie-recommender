```python
import pandas as pd
path = "./dataset/"
credits_df = pd.read_csv(path + "tmdb_5000_credits.csv")
movies_df = pd.read_csv(path + "tmdb_5000_movies.csv")
```

```python
movies_df.head()
```

```python
credits_df.head()
```

```python
from ast import literal_eval
mv_features = ['keywords', 'genres']
for feature in mv_features:
    movies_df[feature] = movies_df[feature].apply(literal_eval)

cd_features = ['cast', 'crew']
for feature in cd_features:
    credits_df[feature] = credits_df[feature].apply(literal_eval)


movies_df.head(10)
```

```python
def get_director(crew):
    for member in crew:
        if member['job'] == 'Director':
            return member['name']
    return None

def get_list(feature, key):
    names = []
    for item in feature:
        names.append(item[key])
    return names
```

```python
movies_df['director'] = credits_df['crew'].apply(get_director)

movies_df['cast'] = credits_df['cast'].apply(lambda x: get_list(x, 'name')[:3])
movies_df['genres'] = movies_df['genres'].apply(lambda x: get_list(x, 'name'))
movies_df['keywords'] = movies_df['keywords'].apply(lambda x: get_list(x, 'name')) 
movies_df.head()
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
        
features = ['cast', 'genres', 'keywords', 'director']
for feature in features:
    movies_df[feature] = movies_df[feature].apply(clean_data)
```

```python
def create_soup(features):
    return ' '.join(features['keywords']) + ' ' + ' '.join(features['cast']) + ' ' + ' '.join(features['genres']) + ' ' + features['director']

movies_df['soup'] = movies_df.apply(create_soup, axis=1)
movies_df[['soup']].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies_df['soup'])
tfidf_matrix.shape
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim.shape
cosine_sim
```

```python
movies = movies_df.reset_index()
indices = pd.Series(movies.index, index=movies['title']).drop_duplicates()
indices.head()
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return movies['title'].iloc[movie_indices]
get_recommendations('The Dark Knight Rises')
```

```python
get_recommendations("Harry Potter and the Chamber of Secrets")
```

```python
def recommend(movie_title):
    for movie in movies_df['title']:
        if movie_title.lower() in movie.lower():
            movie_title = movie
            break
    try:
        recommendations = get_recommendations(movie_title)
        print(f"Because you watched '{movie_title}', you may also like:")
        for i, title in enumerate(recommendations, 1):
            print(f"{i}. {title}")
    except KeyError:
        print(f"Sorry, the movie '{movie_title}' was not found in our database.")
```

```python
movies_export = movies_df[['title', 'soup']]
movies_export.to_csv('movies_soup.csv', index=False)

```
